```python
import cv2

import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread('F HW03-03.jpg')
img_rgb_original = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

imgRGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

height, width, channels = imgRGB.shape

pts1 = np.float32([[55, 115], [115, 103], [68, 143], [134, 126]])
pts2 = np.float32([[30, 140], [95, 140], [30, 170], [95, 170]])

angle = 90

rotation_matrix = cv2.getRotationMatrix2D((width / 2, height / 2), angle, 1)

imgRGB = cv2.warpAffine(imgRGB, rotation_matrix, (width, height))

M = cv2.getPerspectiveTransform(pts1, pts2) 
dst = cv2.warpPerspective(imgRGB, M, (500, 500))

plt.subplot(121), plt.imshow(img_rgb_original), plt.title('Original Image')
plt.axis('off')
plt.subplot(122), plt.imshow(dst), plt.title('Perspective Transformation')
plt.axis('off')
plt.show()
```
